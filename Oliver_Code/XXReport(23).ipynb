```python
import datetime
import Date
import pandas as pd
import numpy as np
import Forecast
import Date
import Demand3
import requests
import matplotlib.pyplot as plt
```

```python
fivehundred = np.zeros([365,40])
sevenfifty = np.zeros([365,40])
onethousand = np.zeros([365,40])

# For TESA
people = 500
for q in range(0,40):
    
    annual = q
    
    for n in range(365):

        daywanted = n+1
        temp1 = Demand3.TESA_HEAT(daywanted, people, annual).sum()[0]
        temp2 = Demand3.TESA_COOL(daywanted, people, annual).sum()[0]

        fivehundred[n,q] = temp1+temp2
        
people = 750
for q in range(0,40):
    
    annual = q
    
    for n in range(365):

        daywanted = n+1
        temp1 = Demand3.TESA_HEAT(daywanted, people, annual).sum()[0]
        temp2 = Demand3.TESA_COOL(daywanted, people, annual).sum()[0]

        sevenfifty[n,q] = temp1+temp2
        
people = 1000
for q in range(0,40):
    
    annual = q
    
    for n in range(365):

        daywanted = n+1
        temp1 = Demand3.TESA_HEAT(daywanted, people, annual).sum()[0]
        temp2 = Demand3.TESA_COOL(daywanted, people, annual).sum()[0]

        onethousand[n,q] = temp1+temp2
```

```python
Afivehundred = fivehundred.sum(axis=0)
Asevenfifty = sevenfifty.sum(axis=0)
Aonethousand = onethousand.sum(axis=0)
```

```python
fig, ax = plt.subplots(figsize=[14,7])

ax.plot(Afivehundred,color='maroon', label='500 Occupants')
ax.plot(Asevenfifty,color='dodgerblue', label='750 Occupants')
ax.plot(Aonethousand,color='teal', label='1000 Occupants')
# ax.set_ylim(350000,630000)
# ax.text(7.5,min(Afivehundred)-8000,'',color='black',fontsize=20)
ax.set_ylabel('Total kWh\n per year',fontsize=15)
ax.set_xlabel('kWh/m\u00B2',fontsize=15)
ax.legend()
ax.grid(lw=0.5)
fig.suptitle('TESA Optimisation',fontsize=20)
```
