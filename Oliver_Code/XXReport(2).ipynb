```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import time
```

```python
import Demand
import Date
```

```python
data = np.zeros((24,31))
data2 = np.zeros((24,31))
Labels = ['\n\nWednesday','\n\nThursday','\n\nFriday','\n\nSaturday','\n\nSunday','\n\nMonday','\n\nTuesday','\n\nWednesday','\n\nThursday','\n\nFriday','\n\nSaturday','\n\nSunday','\n\nMonday','\n\nTuesday','\n\nWednesday','\n\nThursday','\n\nFriday','\n\nSaturday','\n\nSunday','\n\nMonday','\n\nTuesday','\n\nWednesday','\n\nThursday','\n\nFriday','\n\nSaturday','\n\nSunday','\n\nMonday','\n\nTuesday','\n\nWednesday','\n\nThursday','\n\nFriday']

for q in range(31):
    demand = Demand.getGAdemand(q+3)
    demand = demand.drop('00:00.1', axis=0)
    demand2 = Demand.getTESAdemand(q+3)
    demand2 = demand2.drop('00:00.1', axis=0)
    for n in range(0,24):
        data[n,q] = demand.iloc[n]
        data2[n,q] = demand2.iloc[n]

sum = data.sum(axis=0)
sum2 = data2.sum(axis=0)
x = len(sum)
x = np.arange(x)

fig, ax = plt.subplots(figsize=[18,5])

fig.suptitle('January',fontsize=16)

ax.step(x, sum, color='crimson', label='GA daily demand')
ax.step(x, sum2, label='TESA daily demand')
ax.set_ylim(0,1600)
ax.set_xlabel('\n\nDay of the week in January',fontsize=10)
ax.set_ylabel('Total electricity demand (kWh)',fontsize=10)
ax.set_xticks(range(31))
ax.set_xticklabels(Labels, rotation='vertical')

# ax.subplots_adjust(left=0.5)
# for label in ax.xaxis.get_xticklabels():
#     label.set_horizontalalignment('right')

# ax.margins(0.2)

ax.grid()
ax.legend(loc='lower right')
```
