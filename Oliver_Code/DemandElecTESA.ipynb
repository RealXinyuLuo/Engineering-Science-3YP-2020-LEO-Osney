```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
daywanted = 350


def getTESAdemand(daywanted):
    
    rows_to_keep = [0,daywanted]
    
    TESA_elec_demand = pd.read_csv('Data/TESA1718.csv', index_col='Date', skiprows = lambda x: x not in rows_to_keep)
    
    TESA_elec_demand = TESA_elec_demand.T
    
    return TESA_elec_demand
```

```python
TESA_elec_demand = getTESAdemand(daywanted)
```

```python
plt.plot(TESA_elec_demand)
plt.xlabel('Time throughout the day')
plt.ylabel('TESA electricity demand (kWh)')
plt.title('TESA electricity demand (kWh)')
plt.xticks((0,5,11,17,24),('00:00','06:00','12:00','18:00','24:00'))
plt.ylim(0,80)
plt.grid()
plt.show()
```
