```python
import datetime
import Date
import pandas as pd
import numpy as np
import Forecast
import Date
import Demand
import requests
import matplotlib.pyplot as plt
import Price
```

```python
daywanted1 = 2
datewanted1 = Date.getdatewanted(daywanted1, 2014)
raw1 = Demand.getrawHEATdemand(datewanted1)

daywanted2 = 180
datewanted2 = Date.getdatewanted(daywanted2, 2014)
raw2 = Demand.getrawHEATdemand(datewanted2)
```

```python
d = [d for d in range(24)]

fig, ax = plt.subplots(2,sharex=True,sharey=True,figsize=[7,7])

fig.suptitle('Total raw heat demands for the heat network')

ax[0].step(d,raw1,color='crimson')
ax[0].set_title('02 January')
ax[0].set_ylabel('(kW)',fontsize=15)
ax[0].grid()
ax[1].step(d,raw2)
ax[1].set_title('29 June')
ax[1].set_xlabel('Time throughout the day', fontsize=15)
ax[1].set_ylabel('(kW)',fontsize=15)
ax[1].set_xticks((0,4,8,12,16,20,24))
ax[1].set_xticklabels(('00:00','04:00','08:00','12:00','16:00','20:00','24:00'))
ax[1].grid()
```
