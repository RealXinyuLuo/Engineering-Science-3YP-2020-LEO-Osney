```python
import datetime
import Date
import pandas as pd
import numpy as np
import Forecast
import Date
import Demand
import requests
import matplotlib.pyplot as plt
import Price
```

```python
daywanted = 156

price = Price.getvariablebuyprice(daywanted)
demand = Demand.gettotaldemand(daywanted)
Date.getdatewanted(daywanted, 2014)
```

```python
d = np.array([d for d in range(24)])

fig, ax = plt.subplots(2, sharex=True,figsize=[5,7])

ax[0].step(d, price)
ax[0].set_ylim(0,0.25)
ax[0].set_title('Electricity Import Price 05/06/2018')
ax[0].set_ylabel('Price (£)')
ax[0].grid()

ax[1].step(d, demand, color='crimson')
ax[1].set_ylim(0,100)
ax[1].set_title('Total Demand 05/06/2018')
ax[1].set_xlabel('Time throughout the day', fontsize=10)
ax[1].set_ylabel('(kW)',fontsize=10)
ax[1].set_xticks((0,5,11,17,23))
ax[1].set_xticklabels(('00:00','06:00','12:00','18:00','23:00'))
ax[1].grid()
```
