```python
import datetime
import Date
import pandas as pd
import numpy as np
import Forecast
import Date
import Demand
import requests
import matplotlib.pyplot as plt
```

```python
tesajanheat = np.zeros([24,31])
tesajancool = np.zeros([24,31])
tesajulheat = np.zeros([24,31])
tesajulcool = np.zeros([24,31])
gajanheat = np.zeros([24,31])
gajancool = np.zeros([24,31])
gajulheat = np.zeros([24,31])
gajulcool = np.zeros([24,31])
```

```python
for q in range(31):
    print(q)
    daywanted = q+1
    for n in range(24):
        tesajanheat[n,daywanted-1] = Demand.TESA_HEAT(daywanted,750).iloc[n][0]
        tesajancool[n,daywanted-1] = Demand.TESA_COOL(daywanted,750).iloc[n][0]
        tesajulheat[n,daywanted-1] = Demand.TESA_HEAT(daywanted+182,750).iloc[n][0]
        tesajulcool[n,daywanted-1] = Demand.TESA_COOL(daywanted+182,750).iloc[n][0]

        gajanheat[n,daywanted-1] = Demand.GA_HEAT(daywanted).iloc[n][0]
        gajancool[n,daywanted-1] = Demand.GA_COOL(daywanted).iloc[n][0]
        gajulheat[n,daywanted-1] = Demand.GA_HEAT(daywanted+182).iloc[n][0]
        gajulcool[n,daywanted-1] = Demand.GA_COOL(daywanted+182).iloc[n][0]
```

```python
tjanh = tesajanheat.mean(axis=1)
tjanc = tesajancool.mean(axis=1)
tjulh = tesajulheat.mean(axis=1)
tjulc = tesajulcool.mean(axis=1)
gjanh = gajanheat.mean(axis=1)
gjanc = gajancool.mean(axis=1)
gjulh = gajulheat.mean(axis=1)
gjulc = gajulcool.mean(axis=1)

tjanh = np.append(tjanh,tjanh[23])
tjanc = np.append(tjanc,tjanc[23])
tjulh = np.append(tjulh,tjulh[23])
tjulc = np.append(tjulc,tjulc[23])
gjanh = np.append(gjanh,gjanh[23])
gjanc = np.append(gjanc,gjanc[23])
gjulh = np.append(gjulh,gjulh[23])
gjulc = np.append(gjulc,gjulc[23])
```

```python

d = np.array([d for d in range(25)])

fig, ax = plt.subplots(2,2, figsize=[20,7], sharey = True, sharex = True)

ax[0,0].plot(d,tjanh,color='crimson', label='heat')
ax[0,0].plot(d,tjanc,color='dodgerblue', label='cool')
ax[0,0].set_title('TESA - January')
ax[0,0].set_ylabel('kW')
ax[0,0].legend()
ax[0,0].grid(lw=0.5)
ax[0,0].set_xticks((0,3,6,9,12,15,18,21,24))
ax[0,0].set_xticklabels(('00:00','','06:00','','12:00','','18:00','','24:00'))
ax[0,1].plot(d,tjulh, color='crimson', label='heat')
ax[0,1].plot(d,tjulc,color='dodgerblue', label='cool')
ax[0,1].set_title('TESA - July')
ax[0,1].legend()
ax[0,1].grid(lw=0.5)
ax[1,0].set_xlabel('Time')
ax[1,1].set_xlabel('Time')

ax[1,0].plot(d,gjanh,color='crimson', label='heat')
ax[1,0].plot(d,gjanc,color='dodgerblue', label='cool')
ax[1,0].set_title('GA - January')
ax[1,0].legend()
ax[1,0].grid(lw=0.5)
ax[1,0].set_ylabel('kW')
ax[1,1].plot(d,gjulh, color='crimson', label='heat')
ax[1,1].plot(d,gjulc,color='dodgerblue', label='cool')
ax[1,1].set_title('GA - July')
ax[1,1].legend()
ax[1,1].grid(lw=0.5)
```
