```python
__version__ = '0.3'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
import_price1 = pd.read_csv('data/octopus_imprt_price_southern.csv', index_col=0,parse_dates=True,header = 0)
export_price1 = pd.read_csv('data/octopus_export_price_southern.csv', index_col=0,parse_dates=True,header = 0)

start_time = "2019-1-31 00:00:00"
end_time = "2019-1-31 23:30:00"
        
import_price2 =import_price1.loc[start_time:end_time]
export_price2 =export_price1.loc[start_time:end_time]
        
import_price3 = import_price2.between_time('00:00:00','23:30:00')
export_price3 = export_price2.between_time('00:00:00','23:30:00')
        
```

```python
import_price3
```

```python
export_price3
```

```python
for column in import_price3 [['Import Rate (p/kWh)']]:
    import_price = import_price3[column] 
    import_price_value = import_price.values
            
for column in export_price3 [['Outgoing Agile Rate (p/kWh)']]:
    export_price = export_price3[column]
    export_price_value = export_price.values
```

```python
result = pd.concat([import_price3, export_price3], axis=1)
```

```python
export_price3.plot(style='k.',legend=False,use_index=True)
plt.xlabel('')
plt.ylabel("Export Rate ""(p/ kWh)")
plt.title("Variation of Peak-hour Electricity Outgoing Rate with Time")
plt.savefig('Peak-hour Export Rate.png')
plt.show() 
```

```python
plt.figure(1)
result.plot(style='.',legend=True,use_index=True)
plt.xlabel('')
plt.ylabel("Price for Electricity""(p/ kWh)")
plt.title("Variation Energy Price During a Typical Day")
plt.savefig('Day_Plot_Compare.png')
plt.show() 
```

```python
import_price3.plot(style='r',legend=False,use_index=True)
plt.xlabel('')
plt.ylabel("Import Rate ""(p/ kWh)")

plt.title("Variation of Off-Peak Electricity Importing Rate with Time")
plt.savefig('Off-Peak Import Rate.png')
plt.show() 
```
